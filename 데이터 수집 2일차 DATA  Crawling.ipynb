```python
driver = webdriver.Chrome(r"C:\Users\user\chromedriver.exe")
```

```python
driver.get("https://corners.gmarket.co.kr/Bestsellers")
```

```python
driver.maximize_window()
```

```python
from tqdm import tqdm_notebook as tq
```

```python
category_list = []
item_list = []
price_list = []

for i in tq(range(1, 201)):
    #driver.implicitly_wait(n) # 웹 페이지가 완전히 열릴 때까지 n초 동안 기다림
    # 상품 버튼
    try :
        item = driver.find_element_by_xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li[{}]/div[1]/a'.format(i))
        item.click()
        time.sleep(0.2)
        category = driver.find_element_by_css_selector("div.location-navi > ul > li.on > a")
        item_name = driver.find_element_by_css_selector("h1.itemtit")
        item_price = driver.find_element_by_css_selector("strong.price_real")
        category_list.append(category.text)
        item_list.append(item_name.text)
        price_list.append(item_price.text)
        driver.back()
    except : 
        item_name = driver.find_element_by_css_selector("h1.itemtit")
        item_price = driver.find_element_by_css_selector("strong.price_real")
        category_list.append("없음")
        item_list.append(item_name.text)
        price_list.append(item_price.text)
        driver.back()
    
```

```python
len(category_list), len(item_list), len(price_list)
```

```python
data = {"카테고리" : category_list, "품명" : item_list, "가격" : price_list}
```

```python
gmarket_best_list = pd.DataFrame(data)
gmarket_best_list
```

```python
gmarket_best_list.to_excel("지마켓 베스트 상품 리스트.xlsx", encoding = "UTF-8-sig")
```
