```python
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
```

```python
import pickle
```

```python
with open('three_articles.pkl', "rb") as file:
    articles = pickle.load(file)
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
text = articles[0] + articles[1] + articles[2]
```

```python
text = text.replace("\n", "")
text = text.replace("\'", "")
text = text.replace("  ", " ")
text = text.replace("\xa0", "")
text
```

```python
dont = ["것", "거", "때", "개", "안", "여기", "해", "누가", "이", "저", "걸", 
         "은", "는", "로", "나", "의", "로", "를", "것", "거나", "중략",
         "구나", "위해", "게", "대해", "체", "로서", "도", "응", "과", "뭐", 
         "좀", "건가", "기", "것일", "뿐", "리다", "처럼", "마다", "꼭",
         "데", "때문", "수", "더", "단", "명", "틈", "타", "쯤", "씨", "중",
         "등", "위", "사진", "연합뉴스"]
nouns = okt.nouns(" ".join(okt.morphs(text,norm=True,stem=True)))
nouns
```

```python
for i in nouns:
    if i in dont:
        nouns.remove(i)
for i in range(len(nouns)):
    if nouns[i] == "윤":
        nouns[i] = "윤석열"
nouns
```

```python
count = Counter(nouns)
count
```

```python
count.most_common(100)
```

### 워드클라우드 선언하기

```python
wordcloud = WordCloud(font_path="C:\\Windows\\Fonts\\batang.ttc",
                     width = 700,
                     height = 700,
                     background_color = "white",
                     colormap = 'bone',
                     random_state=1).generate_from_frequencies(count)
```

```python
wordcloud
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("뉴스 세 개에서 뽑은 키워드", size = 40,font = "gulim")
plt.show
```
