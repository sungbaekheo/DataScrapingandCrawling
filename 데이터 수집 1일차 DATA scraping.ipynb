```python
res = req.get("https://music.bugs.co.kr/chart")
```

```python
res
```

```python
bugs_chart = res.text
```

```python
bugs_chart = bs(bugs_chart, "html.parser")
bugs_chart
```

```python
ranking = bugs_chart.select("div.ranking>strong")
title = bugs_chart.select("p.title>a")
artist = bugs_chart.select("p.artist>a")

ranking_list = []
title_list = []
artist_list = []

for i in range(len(ranking)):
    ranking_list.append(ranking[i].text)
    title_list.append(title[i].text)
```

```python
artist_check = bugs_chart.select("a.more")
artist_check

for i in artist_check:
    artist.remove(i)

for i in artist:
    artist_list.append(i.text)

len(ranking_list)
len(title_list)
len(artist_list)
data = {"순위" : ranking_list, "곡" : title_list, "아티스트" : artist_list}
```

```python
len(artist_list)
```

```python
artist_list
```

```python
chart = pd.DataFrame(data)
```

```python
chart
```

```python
chart.to_excel("벅스 실시간 인기차트 100.xlsx", encoding = "UTF-8-sig")
```

```python
page = range(1, 11)
```

```python
total_title_list = []
total_review_list = []
total_score_list = []

for i in tqdm_notebook(page):
    url = f"https://movie.naver.com/movie/point/af/list.naver?&page={i}"
    
    res = req.get(url)
    naver_movie = res.text
    movie_reviews = bs(naver_movie, "html.parser")
    
    movie_title = movie_reviews.select("a.movie")
    movie_score = movie_reviews.select("div.list_netizen_score>em")
    movie_review = movie_reviews.select("td.title")

    for j in movie_title:
        total_title_list.append(j.text)
    for j in movie_score:
        total_score_list.append(j.text)
    for i in movie_review:
        test = i.text
        test = test.split("\n\n")[2]
        test = test.replace("\n", "")
        test = test.replace("\t", "")
        test = test.replace("신고", "")
        total_review_list.append(test)

data = {"영화 제목" : total_title_list, "평점" : total_score_list, "리뷰" : total_review_list}

    
    
```

```python
naver_movie_review = pd.DataFrame(data)
```

```python
naver_movie_review
```
