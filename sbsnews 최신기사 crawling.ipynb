```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from tqdm import tqdm_notebook as tq
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://news.sbs.co.kr/news/newsflash.do?plink=NEW&cooper=SBSNEWSMAIN")
```

```python
driver.maximize_window()
```

```python
news_titles = []
times_news = []
news_urls = []
news_views = []
news_articles = []

page_count = 1
while True:
    page_url = "https://news.sbs.co.kr/news/newsflash.do?plink=NEW&cooper=SBSNEWSMAIN&pageIdx={}".format(page_count)
    driver.get(page_url)
    time.sleep(1)
    
    sbsnews = bs(driver.page_source,"html.parser")
    urls = sbsnews.select('a.news')
    
    if len(urls) == 0:
        break
    
    for i in tq(range(len(urls))):
        news_url = "https://news.sbs.co.kr/"+urls[i]['href']
        driver.get(news_url)
        time.sleep(1)
        
        sbsnews = bs(driver.page_source,"html.parser")
        
        title = sbsnews.select_one('#vmNewsTitle')
        times = sbsnews.select_one('div.info_area > span > span:nth-child(2)')
        views = sbsnews.select_one('div.info_area > span > span.view.viewcnt')
        article = sbsnews.select_one('div.main_text > div')
        
        news_titles.append(title.text)
        times_news.append(times.text)
        news_views.append(views.text)
        news_urls.append(news_url)
        news_articles.append(article.text)
        
        driver.back()
        time.sleep(1)
        
        sbsnews = bs(driver.page_source,"html.parser")
    
    page_count += 1
    
```

```python
len(news_titles), len(times_news), len(news_urls), len(news_views), len(news_articles)
```

```python
data = {"기사 제목" : news_titles, "기사 내용" : news_articles, "기사 URL" : news_urls}
```

```python
latest_news_list = pd.DataFrame(data)
```
