```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from tqdm import tqdm_notebook as tq
```

```python
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://www.youtube.com/channel/UCkuA_gDjISfGgbdp02BUwyQ/videos")
```

```python
driver.maximize_window()
```

```python
from bs4 import BeautifulSoup as bs
```

```python
title = driver.find_element_by_css_selector("#video-title")
view = driver.find_element_by_css_selector("#metadata-line > span:nth-child(1)")
```

```python
body = driver.find_element_by_css_selector("body")

while True :
    last = bs(driver.page_source, "html.parser")
    for i in range(20):
        body.send_keys(Keys.END)
    new = bs(driver.page_source, "html.parser")
    if last == new:
        break
    else :
        pass
```

```python
title = driver.find_elements_by_css_selector("#video-title")
view = driver.find_elements_by_css_selector("#metadata-line > span:nth-child(1)")

video_titles = []
video_views = []

for i in tq(title):
    print(i.text)
    video_titles.append(i.text)
```

```python
video_views = []
for i in tq(view):
    string = i.text
    string = string.replace("조회수", "")
    string = string.replace("회", "")
    string = string.replace(".", "")
    string = string.replace("천", "000")
    string = string.replace("만", "0000")
    string = int(string)
    video_views.append(string)
    print(string)
```

```python
len(video_views)
```

```python
data = {"영상 제목" : video_titles, "조회수" : video_views}
```

```python
kjipsa = pd.DataFrame(data)
```

```python
kjipsa.sort_values(by=["조회수"], ascending = False)
```
